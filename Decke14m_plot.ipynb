```python
"""
created by
Tobias Hein
29.11.2018
"""

##########
#Datenstruktur:
#
#Kennung
#X
#Y
#SF (sofern vorhanden)
#
##########

##########
#execfile("C:\\he\python\\Decke14m_plot\\Dacke14m_plot.py")
##########

#Import
import csv
import matplotlib
import matplotlib.pyplot as plt
from scipy.misc import imread
from matplotlib.backends.backend_pdf import PdfPages
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

#Definitionen
path="C:\\he\python\\Decke14m_plot\\dataDM_(neu)_181129.csv"
plotbackground="C:\\he\\python\\Decke14m_plot\\Decke14m.png"
FSIZE=10		#Fontsize von Titel und Achsenbeschriftung

def string2float(value):
    value = value.replace('.','')
    value = value.replace(',','.')
    return float(value)

#Daten einlesen
csvdata=[]
ofile=open(path, 'r')
Reader = csv.reader(ofile, delimiter=';', dialect='excel')
for row in Reader:
    csvdata.append(row)

ofile.close()

#Daten einsortieren
csv_Kennung=[]
csv_X=[]
csv_Y=[]
csv_SF=[]
KennungPlot=[]
for i in range(1,len(csvdata)):
    csv_Kennung.append(csvdata[i][0])
    csv_X.append(string2float(csvdata[i][1]))
    csv_Y.append(string2float(csvdata[i][2]))
    csv_SF.append(csvdata[i][3])
    KennungPlot.append(csvdata[i][0][5]+csvdata[i][0][7:11]+csvdata[i][0][-3:])

#Daten plotten
fig=plt.figure(num=1, figsize=(30, 16), dpi=80)
ax = fig.add_subplot(1,1,1)
matplotlib.rcParams.update({'font.size': FSIZE})
im1=ax.scatter(csv_Y, csv_X, marker='o', color='red', s=50)
#Background Karte
img = imread(plotbackground)
plt.imshow(img, zorder=0, extent=[-430, 8830, -490, 4500])
#plt.imshow(img, zorder=0, extent=[0, 8300, 0, 3950])
plt.show()

#Plot customization
ax.set_xlabel('Y [cm]')
ax.set_ylabel('X [cm]')
ax.set_title('Deckenmessungen MH 14m')
#plt.ylim(0,3950)
#plt.xlim(0,8300)
plt.ylim(-550,4500)
plt.xlim(-570,8950)
ax.set_aspect('equal')
#for label, x, y in zip(KennungPlot, csv_Y, csv_X):
#    plt.annotate(label,xy = (x, y),xytext = (0, 10),textcoords = 'offset points',ha = 'center', va = 'center',fontsize=7)
plt.show()

#Plot als PDF speichern
pdf = PdfPages(path[0:-4]+'_TEST.pdf')
pdf.savefig(fig)
pdf.close()

#Plot schliessen
plt.close('all')
```
